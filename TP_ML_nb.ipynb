```python
def count_words(texts):
    """Vectorize text : return count of each word in the text snippets

    Parameters
    ----------
    texts : list of str
        The texts

    Returns
    -------
    vocabulary : dict
        A dictionary that points to an index in counts for each word.
    counts : ndarray, shape (n_samples, n_features)
        The counts of each word in each text.
        n_samples == number of documents.
        n_features == number of words in vocabulary.
    """
    words = set()
    vocabulary = {}
    
    for text in texts:
        for word in text.split():
            if word not in vocabulary:
                vocabulary[word] = 1
            else:
                vocabulary[word] += 1

    words = vocabulary.keys()

    n_features = len(words)
    counts = np.zeros((len(texts), n_features))
    
    for i, text in enumerate(texts):
        vocabulary_per_text = dict()
        
        for word in text.split():
            
            if word not in vocabulary_per_text:
                vocabulary_per_text[word] = 1
            else:
                vocabulary_per_text[word] += 1
    
        for j, n_features in enumerate(words):
            if n_features in vocabulary_per_text:
                counts[i][j] = vocabulary_per_text[n_features]
            else:
                counts[i][j] = 0
            
    print(counts)
    print(counts.shape)
    #np.savetxt("fooooo.csv", counts, delimiter=",")
    return vocabulary, counts
```

```python
class NB(BaseEstimator, ClassifierMixin):
    def __init__(self):
        pass

    def fit(self, X, y):
        return self

    def predict(self, X):
        return (np.random.randn(len(X)) > 0).astype(np.int)

    def score(self, X, y):
        return np.mean(self.predict(X) == y)
```
